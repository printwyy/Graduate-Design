```python
d1=pd.DataFrame.from_dict(res1)
d1=d1.set_index(['time'])
print(d1)
```

```python
datadiff1 = d1.diff(1).diff(1).dropna() #  
datadiff1.plot(kind="line",figsize=(12,6))  #  
# plt.grid("on") 
# plt.title("diff data") 
# plt.show()
import statsmodels.api as sm
import numpy as np
lags = [4,8,12,16,20,24,28,32,36,64] 
LB = sm.stats.diagnostic.acorr_ljungbox(datadiff1,lags=lags) 
# LB = pd.DataFrame(data = np.array(LB).T,columns=["LBvalue","P-value"]) 
# LB["lags"]=lags 
print("\n",LB)
fig = plt.figure(figsize=(10,5))   #
ax1 = fig.add_subplot(211) 
fig = sm.graphics.tsa.plot_acf(datadiff1,lags=50,ax=ax1) 
ax2 = fig.add_subplot(212) 
fig = sm.graphics.tsa.plot_pacf(datadiff1,lags=50,ax=ax2,method='ywm') 
plt.subplots_adjust(hspace=0.3) 
plt.show()
```

```python
# 
dftest = sm.tsa.adfuller(datadiff1,autolag="AIC") 
print("adf:",dftest[0]) 
print("pvalue:",dftest[1])
from statsmodels.tsa.stattools import arma_order_select_ic
res =arma_order_select_ic(datadiff1, max_ar=10, max_ma=10, ic=['aic', 'bic', 'hqic'])
print ('AIC-order: {}' .format(res.aic_min_order))
print ('BIC-order: {}' .format(res.bic_min_order))
print ('HQIC-order: {}' .format(res.hqic_min_order)) 
```

```python
# import statsmodels.api as sm
# from statsmodels.tsa.arima_model import ARMA
# model = ARMA(data_diff, order=bic_min_order).fit(disp=-1)
l=len(d1)
ARMAmol = sm.tsa.arima.ARIMA(d1, order=(1,2,2)).fit()  #(AR,0,MA)
predicet_ARMA = ARMAmol.predict(start=0,end=l+15) 
pre=ARMAmol.forecast(15)
print(pre)

plt.figure(figsize=(12,8)) 
orig = plt.plot(d1,color="blue",label="Original") 
predict = plt.plot(predicet_ARMA,color="red",label="Predict") 
plt.title("ARMA(1,1)") 
plt.grid("on") 
plt.legend(loc="upper left") 
plt.show()

```
